```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve, average_precision_score
from ultralytics import YOLO
import os

# Path to the trained model
model_path = '/content/gdrive/My Drive/ColabNotebooks/Results/detect/train2/weights/best.pt'

# Load the trained model
model = YOLO(model_path)
print("Model classes:", model.names)

# Path to the image for predictions
image_path = '/content/gdrive/My Drive/ColabNotebooks/Test.jpg'

# Custom folder to save results
result_folder = '/content/gdrive/MyDrive/ColabNotebooks/Results/predict/runs/detect/predict'
os.makedirs(result_folder, exist_ok=True)

# Define thresholds to test
thresholds = np.linspace(0.1, 1.0, 10)  # Thresholds from 0.1 to 1.0

# Placeholder for precision-recall values
all_precisions = []
all_recalls = []
average_precisions = []

# Ground truth labels for testing (update with your actual dataset)
# Replace with the actual binary ground truth labels for evaluation
# For example: [1 if object is present else 0 for all images]
y_true = [1, 0, 1, 1, 0]  # Example ground truth (binary labels)

for conf in thresholds:
    print(f"Running prediction with confidence threshold: {conf:.2f}")

    # Perform predictions
    results = model.predict(source=image_path, save=True, conf=conf, save_dir=result_folder)

    # Debugging: Check predictions
    y_scores = []  # Predicted confidence scores
    for result in results:
        if len(result.boxes) == 0:
            print(f"No predictions found for threshold {conf:.2f}")
        for pred in result.boxes:
            y_scores.append(pred.conf.item())  # Convert tensor to Python float
            print(f"Predicted Class: {int(pred.cls.item())}, Confidence: {pred.conf.item():.2f}, Coordinates: {pred.xyxy.tolist()}")

    if not y_scores:
        print(f"No detections for threshold {conf:.2f}")
        all_precisions.append([0])
        all_recalls.append([0])
        average_precisions.append(0)
        continue

    # Calculate precision, recall, and average precision
    precision, recall, _ = precision_recall_curve(y_true[:len(y_scores)], y_scores)
    all_precisions.append(precision)
    all_recalls.append(recall)
    average_precisions.append(average_precision_score(y_true[:len(y_scores)], y_scores))

# Plot Precision-Recall curves
plt.figure(figsize=(10, 8))
for i, conf in enumerate(thresholds):
    if len(all_precisions[i]) > 1:  # Skip thresholds with no detections
        plt.plot(all_recalls[i], all_precisions[i], label=f'Threshold: {conf:.2f} (AP={average_precisions[i]:.2f})')

plt.title('Precision-Recall Curves at Different Thresholds')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
